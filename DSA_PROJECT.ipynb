```python
# Visualize the distribution of the target variable
sns.countplot(x='is_safe', data=data)
plt.title('Distribution of Water Safety')
plt.show()
```

```python
# Visualize the correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Visualize the distribution of each numerical feature
numerical_features = data.select_dtypes(include=['float64']).columns
data[numerical_features].hist(bins=20, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Features', y=1.02)
plt.show()
```

```python
# Box plots for each numerical feature
plt.figure(figsize=(15, 10))
sns.boxplot(data=data[numerical_features], orient="h")
plt.title('Box Plots of Numerical Features')
plt.show()
```
